```python
def remove_bad_feat(data, loaded_model, name):#best_acc,
    X = data.loc[:,'Loaded_Language':'Thought-terminating_Cliches',]
    Y = data.loc[:,"BCandidate"]
    acc = []
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,
                                                        test_size = .2,
                                                        shuffle=True)
    
    for i in data.X_train.columns:
        new_X_train = X_train.drop(columns = i)
        new_X_test = X_test.drop(columns= i)
        ##model_training
        
        
        if name == "extra trees":
            MOD = ExtraTreesClassifier(n_estimators =loaded_model.best_params_['n_estimators'],random_state = loaded_model.best_params_['random_state'] )
        elif name == "random forest":
            MOD = RandomForestClassifier(n_estimators =loaded_model.best_params_['n_estimators'],
                                                  random_state = loaded_model.best_params_['random_state'],
                                                  max_depth = loaded_model.best_params_['max_depth'])
        else:
            print("there is a problem with name param")
            return
        MOD.fit(new_X_train, Y_train)    
        
        acc.append(accuracy_score(Y_test,loaded_model.predict(new_X_test)))
    print(max(acc))
    index = acc.index(max(acc))
    worst_feat = data.X_train.columns[index]
    return worst_feat
        
```
